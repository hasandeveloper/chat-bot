```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters.character import CharacterTextSplitter
from langchain_openai.embeddings import OpenAIEmbeddings
from langchain_chroma import Chroma
import re
import copy
```

```python
loader_pdf = PyPDFLoader("SERVICE _ WARRANTY POLICY.pdf")
pages_pdf = loader_pdf.load()
```

```python
pages_pdf
```

```python
# Deep copy to avoid modifying original
pages_pdf_clean = copy.deepcopy(pages_pdf)

# Clean each page
for doc in pages_pdf_clean:
    text = doc.page_content
    
    # 1. removes all extra spaces and line breaks. So your page text becomes a clean, continuous paragraph.
    text = " ".join(text.split())

    # Update the document
    doc.page_content = text

# Example: view the cleaned first page
print(pages_pdf_clean[0].page_content)
```

```python
# split into chunks
char_splitter = CharacterTextSplitter(chunk_size = 500, 
                                      chunk_overlap = 50)
pages_char_split = char_splitter.split_documents(pages_pdf_clean)
print(pages_char_split[1])
```

```python
# Generate embeddings
embeddings_model = OpenAIEmbeddings(model="text-embedding-3-small")  # or 3-large

# Store in Chroma vector DB
vectorstore = Chroma.from_documents(
    documents=pages_char_split,
    embedding=embeddings_model,
    persist_directory="./chroma_db"  # folder to store vectors
)

print(f"Stored {len(pages_char_split)} chunks in Chroma!")
```

```python
vectorstore = Chroma(
    embedding_function=embeddings_model,
    persist_directory="./chroma_db"
)
results = vectorstore.similarity_search("warranty duration for electronics", k=3)
for r in results:
    print(r.page_content)
```
